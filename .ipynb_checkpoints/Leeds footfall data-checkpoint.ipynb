# Leeds footfall data

The [Leeds Data Mill](http://www.leedsdatamill.org) website has footfall data, that is the number of people passing a given location, for eight different locations in Leeds city centre. The figures date back to August 2008 and are recorded automatically with CCTV cameras set up in the various locations.

The data are available to download and are in CSV files, although the earlier data are in a slightly different format to the more recent data (there were separate files for each location and the columns and column headers in the tables have changed).

The data are recorded hourly, 24 hours a day, 365 days a year.

## Reading in the data

The following code reads all of the data into a Pandas dataframe and then shows us the first 10 lines in the table.

```python
import pandas as pd

all_data = pd.read_csv('cleaned_data.csv')
del all_data['Unnamed: 0']
all_data.head(16)
```

## Using the data to answer some questions

Using Pandas' sophisticated grouping, filtering and sorting functions, we can answer some interesting questions about the data.

## What was the percentage change in footfall for the eight locations between 2012 and 2013?

We can create a function to take the dataframe with our footfall data and two years and calculate the percentage change in footfall between those years.

```python
def year_delta(df, year1, year2):
	gp = df.groupby(['BRCYear', 'LocationName'])['InCount'].mean().reset_index()

	df_yr1 = gp.copy()[gp['BRCYear'] == year1]
	df_yr1.rename(columns={'InCount': str(year1) + '_Mean'}, inplace=True)
	del df_yr1['BRCYear']

	df_yr2 = gp.copy()[gp['BRCYear'] == year2]
	df_yr2.rename(columns={'InCount': str(year2) + '_Mean'}, inplace=True)
	del df_yr2['BRCYear']
	
	df_delta = pd.merge(df_yr1, df_yr2, on='LocationName')
	df_delta[str(year2) + '_' + str(year1) + '_Delta'] = ((df_delta[str(year2) + '_Mean'] - df_delta[str(year1) + '_Mean']) / df_delta[str(year1) + '_Mean']) * 100

	return df_delta

delta_2013_2012 = year_delta(all_data, 2012, 2013)
delta_2013_2012.sort(columns='2013_2012_Delta', ascending=False)
```

## Plotting the average hourly footfall through the year

```python
from ggplot import *

by_month = all_data.groupby('BRCMonthName')['InCount'].mean().reset_index()
sorter = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sorter_index = dict(zip(sorter, range(1,len(sorter)+1)))
by_month['MonthRank'] = by_month['BRCMonthName'].map(sorter_index)
by_month.sort('MonthRank', ascending=True, inplace=True)

ggplot(aes(x='MonthRank', y='InCount'), data=by_month) + geom_point() + xlab('Month') + ylab('Average hourly footfall') \
+ xlim(0,13) + scale_x_continuous(breaks=range(1,13), labels=[m[0] for m in sorter])
```
